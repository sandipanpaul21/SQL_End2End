**Q1.** Get all orders where the customer's name has "a" as the second character and "d" as the fourth character.

```python
%%sql
SELECT TOP 2 *
FROM orders
WHERE customer_name LIKE '_a_d%'
```

**Q2.** Get all orders placed in the month of December 2020.

```python
%%sql
SELECT TOP 2 *
FROM orders
WHERE MONTH(order_date) = 12 AND YEAR(order_date) = 2020
```

**Q3.** Get all orders where ship_mode is neither 'Standard Class' nor 'First Class' and ship_date is after November 2020.

```python
%%sql
SELECT TOP 2 *
FROM orders
WHERE ship_mode NOT IN ('Standard Class','First Class') AND ship_date >= '2020-12-01'
```

**Q4.** Get all orders where the customer name neither starts with "A" nor ends with "n".

```python
%%sql
SELECT TOP 2 *
FROM orders
WHERE customer_name NOT LIKE 'A%' AND customer_name NOT LIKE '%n'
```

**Q5.** Get all orders where profit is negative.

```python
%%sql
SELECT TOP 2* 
FROM orders 
WHERE profit < 0;
```

**Q6.** Get all orders where either quantity is less than 3 or profit is 0.

```python
%%sql
SELECT TOP 2 * 
FROM orders 
WHERE quantity < 3 OR profit = 0;
```

**Q7.** Get all orders from the South region where some discount is provided to the customer.

```python
%%sql
SELECT TOP 2 * 
FROM orders 
WHERE region = 'South' AND discount > 0;
```

**Q8.** Find the top 3 orders with the highest sales in the Furniture category.

```python
%%sql
SELECT TOP 3 order_id, ROUND(SUM(sales),1) AS total_sales
FROM orders 
WHERE category = 'Furniture'
GROUP BY order_id
ORDER BY SUM(sales) DESC
```

**Q9.** Find all the records in the Technology and Furniture categories for orders placed in the year 2020 only. 

```python
%%sql
SELECT TOP 2 *
FROM orders
WHERE category IN ('Technology', 'Furniture') AND YEAR(order_date) = 2020
```

**Q10.** Find all the orders where order date is in year 2020 but ship date is in year 2021.

```python
%%sql
SELECT TOP 2 *
FROM orders
WHERE YEAR(order_date) = 2020 AND YEAR(ship_date) = 2021;
```

**Q11.** Find employees whose salary is greater than the average salary of their own department.

```python
%%sql
WITH department_average_salary AS 
(
    SELECT dept_id, AVG(salary) AS dept_avg_salary
    FROM employee
    GROUP BY dept_id
)
SELECT *
FROM employee AS t1
LEFT JOIN department_average_salary AS t2
ON t1.dept_id = t2.dept_id
WHERE t1.salary > t2.dept_avg_salary
```

**Q12.** Find duplicate employee names in a table.

```python
%%sql
SELECT emp_name, COUNT(*) AS name_count
FROM employee t1
GROUP BY emp_name
```

**Q13.** Find Los Angeles orders where the order amount is greater than the city's average order amount.

```python
%%sql
SELECT TOP 3 *
FROM orders
WHERE city = 'Los Angeles' 
AND sales > 
    (
        SELECT AVG(sales) 
        FROM orders
        WHERE city = 'Los Angeles'
    )

```

**Q14.** Find the top 3 highest-paid employees.

```python
%%sql
SELECT TOP 3 *
FROM employee
ORDER BY salary DESC;
```
